```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
# create an empty canvas
ax = plt.axes(projection='3d')
```

```python
# divide the interval <-10,10> into 500 parts
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
 
# a Cartesian product
X,Y = np.meshgrid(x,y)
```

```python
# the parameters of the distribution
# x
mu_x = 0
variance_x = 10
 
# y
mu_y = 0
variance_y = 10
 
# a means list
means = [mu_x, mu_y]
 
# a covariance
cov = [[variance_x, 0], [0, variance_y]]
 
# a multivariate normal distribution
mn = multivariate_normal(means, cov)
```

```python
cov
```

```python
mn
```

```python
# a position matrix
position = np.array([X,Y]).T
 
# z-coordinates from a probability-density function
Z = mn.pdf(position)
```

```python
#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))
 
# set a 3D projection
ax = plt.axes(projection='3d')
 
# a wireframe
ax.plot_wireframe(X, Y, Z, color='r')
plt.show()
```

```python
#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))
 
# set a 3D projection
ax = plt.axes(projection='3d')
 
# the surface
ax.plot_surface(X, Y, Z, cmap='jet')
plt.show()
```

```python
#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))
 
# set a 3D projection
ax = plt.axes(projection='3d')
 
# the surface
ax.contour(X, Y, Z, 40, cmap='jet')
plt.show()
```
