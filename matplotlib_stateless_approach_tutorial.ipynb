```python
# import numpy
import numpy as np
 
# import matplotlib
import matplotlib.pyplot as plt
```

```python
# 200 values from the interval <0,100>, equidistantly divided
x = np.linspace(0,100,200)
y = np.sin(x)
 
# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize = (10,6))
 
# plot the line
ax.plot(x,y,'red')
plt.show()
```

```python
# 200 random values from the interval <0,10>
x = 10*np.random.rand(200,1)
 
# 200 random values from the interval <0,15>
y = 15*np.random.rand(200,1)
 
# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize = (10,6))
 
# scatter plot
ax.scatter(x,y)
plt.show()
```

```python
# 200 values from the interval <0,100>, equidistantly divided
x = np.linspace(0,100,200)
 
# sin(x) values
y1 = np.sin(x)
 
# sin(x)*cos(x) values
y2 =(np.sin(x))*(np.cos(x))
 
# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize = (10,6))
 
# sin(x) plot
ax.plot(x,y1,'red')
 
# sin(x)*cos(x) plpt
ax.plot(x,y2,'blue')
plt.show()
```

```python
# create the figure and axes, set the figure size
fig, axes = plt.subplots(nrows=2, ncols=1 ,figsize = (10,6))
 
# the first plot
axes[0].plot(x,y1,'red')
 
# the first plot title
axes[0].set_title('sin(x)')
 
# the second plot
axes[1].plot(x,y2,'blue')
 
# the second plot title
axes[1].set_title('sin(x)*(cos(x))')
 
# automatically adjust the subplot parameters to give specified padding
fig.tight_layout()
plt.show()
```

```python
# import pandas
import pandas as pd
 
# import sklearn datasets
from sklearn import datasets
```

```python
# load the iris dataset
iris = datasets.load_iris()
 
# create a dataframe
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
 
# create a target
iris_df['target'] = iris.target
 
# map target values to target names
iris_df['target_name'] =iris_df.target.map(
   {0: 'setosa',
    1: 'versicolor',
    2: 'virginica'}
)
```

```python
iris_df.head()
```

```python
# Iris setosa
setosa = iris_df[iris_df.target_name == 'setosa']
 
# Iris versicolor
versicolor = iris_df[iris_df.target_name == 'versicolor']
 
# Iris virginica
virginica = iris_df[iris_df.target_name == 'virginica']
 
# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize=(10,6))
 
# scatter the plot of setosa
ax.scatter(setosa['sepal length (cm)'],setosa['sepal width (cm)'],
           marker ='o', color = 'red', label = 'setosa')
 
 
# scatter the plot of versicolor
ax.scatter(versicolor['sepal length (cm)'],versicolor['sepal width (cm)'],
           marker ='o', color = 'green', label = 'versicolor')
 
 
# scatter the plot of virginica
ax.scatter(virginica['sepal length (cm)'],virginica['sepal width (cm)'],
           marker ='o', color = 'blue', label = 'virginica')
 
# set the legend
ax.legend(loc='upper right')
 
# set the title
ax.set_title('Iris flower')
 
# set the x-label
ax.set_xlabel('sepal length (cm)')
 
# set the y-label
ax.set_ylabel('sepal width (cm)')
plt.show()
```
